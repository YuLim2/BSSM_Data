&#9997; **Note: Python 리스트(list)**

---
+ 리스트(list)
 - 리스트의 각 요소는 순서를 가지고 있음
 - 대괄호([])로 작성되며 각 요소는 쉼표(,)로 구분
 - 요소를 추가, 수정, 삭제 가능

+ 리스트 접근 
 - for ... in 
 
+ 조건에 맞는 리스트 출력 추출
  > 
  ```
  [item for item in lt]
  [item for item in lt if 조건식]
  [item if 조건식 else 조건식 for item in lt]
  ``` 


````python
#열인덱스를 리스트로 만들기
lt = list(df.columns)
````

````python
#리스트 한줄로 항목 출력 
for item in lt:
  print(item) 
````

````python
#리스트 한줄로 조건에 맞는 항목 출력 
[item for item in lt if item == '년도']
````

````python
#리스트 한줄로 조건에 맞는 항목 출력 
lt = [item if item == '년도' else item[:-3] for item in lt]
````

````python
#데이터프레임 열명리스트 만들기 
df.columns = lt
````

````python
#데이터프레임 열명 출력
df.columns 
````

````python
#데이터프레임 변경 확인
df.info()
````

## **데이터프레임 추출하기**
+ 열자료 가져오기 : df[]
 + 특정 열 가져오기 : df[열인덱스], df[열인덱스 리스트]
 + 특정 행 가져오기 : df[시작행인덱스:종료행인덱스] 

+ 인덱스명으로 특정 항목 가져오기 : df.loc[행인덱스, 열인덱스]
  * 특정 열 가져오기 : df.loc[:, 열인덱스]
  * 특정 행 가져오기 : df.loc[행인덱스, :]
  * 특정 항목 가져오기 : df.loc[행인덱스, 열인덱스]

+ 인덱스 위치값으로 특정 항목 가져오기 : df.iloc[행인덱스순서, 열인덱스순서]
  * 특정 열 가져오기 : df.iloc[:, 열인덱스순서]
  * 특정 행 가져오기 : df.iloc[행인덱스순서, :]
  * 특정 항목 가져오기 : df.iloc[행인덱스순서, 열인덱스순서]

### loc = 이름을 가지고 추출
### iloc = 위치를 가지고 추출

````python
#한 개의 열 자료 가져오기
df1 = df.loc[1971:1972, '출생아수':'사망자수']
````

````python
#여러 개의 열 자료 가져오기
df.loc[1971:1973,'출생아수':]
````

````python
#한 개의 열 자료 데이터프레임으로 가져오기
df.loc[1971:1972, '출생아수':'사망자수']
````

````python
#특정 자료 가져오기
df.loc[2015:2017, '출생아수':'사망자수']
````

## 데이터프레임 새로운 열 생성
+ df['새로운열명'] = 시리즈데이터

````python
#새로운 열 생성
df['자연증가수'] = df['출생아수'] - df['사망자수']
````

````python
df.info()
````

````python
#전년도 출생아수열 만들기
lt = list(df['출생아수']) 
lt.insert(0, 0)
lt.pop()
df['전년출생아수'] = lt
df.info()
````

````python
# 전년 대비 출생아수
df['전년대비출생아수'] = df['출생아수'] - lt
df.info()
````

## **조건에 맞는 자료 추출**


---
+ df[조건식]
  + 시리즈.isin([리스트])
  + and 연산 : &
  + or 연산 : | 
  + and 연산이나 or연산인 경우는 반드시 ()로 묶어서 사용

````python
#전년도 대비 출생아수가 감소한 년도 
lt = list(df['전년대비출생아수'])
lt
````

````python
lt = ['증가' if item >= 0 else '감소' for item in lt]
lt
````

````python
#전년대비 증가 감소 구분 
df['증감'] = lt
df.info()
````
